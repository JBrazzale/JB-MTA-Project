```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd


```

```python
engine = create_engine("sqlite:///mta_data.db")
```

```python
insp = inspect(engine)
print(insp.get_table_names())
#this will allow us to see our table name(s)
```

```python
full_mta_data = pd.read_sql('SELECT * FROM mta_data;', engine)
full_mta_data 
#This will give us an outline of what our dataframe looks like
```

```python
no_of_stations = pd.read_sql('SELECT COUNT(DISTINCT STATION) AS "Number of Stations" FROM mta_data;', engine)
no_of_stations
#quick check to see the number of unique station names
```

```python
unique_station_units = pd.read_sql('SELECT STATION, COUNT(DISTINCT UNIT) FROM mta_data GROUP BY STATION ORDER BY COUNT(DISTINCT UNIT) DESC LIMIT 5;', engine)
unique_station_units
#query to see which station has the highest number of unique units
```
